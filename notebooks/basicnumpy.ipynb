```python
#2. Conditional Replacement
#Create array from 1 to 20.
#Replace all numbers:
#divisible by 3 → -1
#divisible by 5 → 100
#(Do it using NumPy, not loops.)

import numpy as np
arr=np.arange(1,21)
print(arr)
a=np.where(arr%3==0,-1,arr)
a=np.where(arr%5==0,100,arr)
print(a)
```
