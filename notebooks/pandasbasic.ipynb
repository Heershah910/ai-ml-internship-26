```python
#MATHEMATICAL OPERATION ON 2 SERIES
import pandas as pd

match1 = pd.Series([200,300,200,100,500],
                   index=["a","b","c","d","e"])

match2 = pd.Series([200,300,200,100,500],
                   index=["x","b","c","d","e"])
#add
print(match1 + match2)

result = match1.add(match2, fill_value=0)
print(result)

#sub
print(match1-match2)
result=match1.sub(match2)
print(result)


#mul
print(match1*match2)
result=match1.mul(match2)
print(result)

#div
print(match1/match2)
result=match1.div(match2)
print(result)
```

```python
# STATISTICALLY METHOD

s=pd.Series([10,20,30,40,50])
print(s)
print(s.sum())
print(s.mean())
print(s.max())
print(s.min())
print(s.median())
print(s.var())
print(s.std())
print(s.mode())
print(s.mode())
```

```python
# BASIC DATA CLEANING METHOD

s=pd.Series([10,20,30,40,50])
print(s)
print(s.isnull())

s=pd.Series([10,20,None,40,50])
print(s.isnull())

s=pd.Series([10,20,None,40,50])
print(s.isnull().sum())  # IT WILL COUNT HOW MANY DATA ARE NULL

s=s.fillna(0)  #IF SEQUENCE SHOULD BE MAINTAINED THEN USE FILLNA FUNCTION
print(s)

d=s.dropna()  #RECORDS WITH NAN WILL BE DELETED USE DROPNA FUNCTION
print(d)
```
